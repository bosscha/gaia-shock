### Upgrade fields in list of SCs

From the SC list, new fields are added to help with future analysis: votname, etc...

```julia
using DataFrames , CSV , Statistics , Printf
using PyPlot

## directory
rootdir = "/home/stephane/Science/cluster/GAIA"
datadir = "$rootdir/master/notebooks/data"
wdir    = "/home/stephane/Science/cluster/GAIA/products"
plotdir = "/home/stephane/Science/cluster/GAIA/products/testfull"

cd(wdir)
```

```julia
## read the SC parameter file

filelist = "$datadir/BrowseTargets.18292.1530479692.fake"
sclist = CSV.read(filelist, delim = "|")
delete!(sclist,1)
delete!(sclist,11)
```

```julia
## Adding new columns
## Check the value of radius in getVotable if the definition changed for the downloading

sclist[:votname] = map((x,y) -> @sprintf("%s-%3.1fdeg.vot", strip(x), 5*float(y)) , sclist[:name], sclist[:cluster_radius])

CSV.write("mwsc-vot.csv", sclist, delim =";")
```
