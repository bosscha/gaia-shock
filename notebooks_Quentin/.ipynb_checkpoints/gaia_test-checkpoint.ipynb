```python
cluster_file = open("BrowseTargets.18292.1530479692.gaia.selected.txt", "r")
Stop = False
line = cluster_file.readline()
i = 0
clusters = []
while not Stop :
    i += 1
    line = cluster_file.readline()
    if line == "" : break
    Z = line.split('|')
    ncluster = int(Z[7])
    dist = int(Z[8])
    name = Z[1][:Z[1].find(" ",1+Z[1].find(" "))]
    
    if ncluster > 1000 :
        clusters.append((name,ncluster,dist))
    
    #if i == 30 : Stop = True

for c in clusters :
    print(c)
```
